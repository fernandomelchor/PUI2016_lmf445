```python

md['chisq'] = {} 

#and do it in a for loop. not pythonic, but easily readable

for n in mysize:
    md['chisq'][n] = np.random.chisquare(df, size = n)
    
# save the means for plotting later    

md['chisq']['means'] = {}
axchisq_mu_n = pl.figure(figsize=(10,6)).add_subplot(111)

for nn in md['chisq']:
    if not type(nn) == str:
        md['chisq']['means'][nn] = md['chisq'][nn].mean()
        #and plot it
        axchisq_mu_n.plot(nn, md['chisq']['means'][nn], 'o')
        axchisq_mu_n.set_xlabel('sample size', fontsize=18)
        axchisq_mu_n.set_ylabel('sample mean', fontsize=18)
        axchisq_mu_n.set_title('Chi squared', fontsize=18)
        axchisq_mu_n.plot([min(mysize), max(mysize) + 10], [df, df], 'k')
```

```python
#The bigger the sample the deviation from the mean is smaller
chimean = list(md['chisq']['means'].values())

pl.figure(figsize=(10, 10))
pl.hist(chimean,bins=30)
pl.xlabel('sample mean', fontsize = 18)
pl.ylabel('N', fontsize = 18)
pl.suptitle("Chi-squared mean",fontsize=18)
```

```python
md['Normal'] = {} 
 
#and do it in a for loop. not pythonic, but easily readable
n = 0

for n in mysize:
    md['Normal'][n] = np.random.normal(df,5,n)
    
# save the means for plotting later    

md['Normal']['means'] = {}
axnormal_mu_n = pl.figure(figsize=(10,6)).add_subplot(111)

for nn in md['Normal']:
    if not type(nn) == str:
        md['Normal']['means'][nn] = md['Normal'][nn].mean()
        #and plot it
        axnormal_mu_n.plot(nn, md['Normal']['means'][nn], 'o')
        axnormal_mu_n.set_xlabel('sample size', fontsize=18)
        axnormal_mu_n.set_ylabel('sample mean', fontsize=18)
        axnormal_mu_n.set_title('Normal Distribution', fontsize=18)
        axnormal_mu_n.plot([min(mysize), max(mysize) + 10], [df, df], 'k')
```

```python
#The bigger the sample the deviation from the mean is smaller
normean = list(md['Normal']['means'].values())

pl.figure(figsize = (10, 10))
pl.hist(normean,bins = 30)
pl.xlabel('sample mean', fontsize = 18)
pl.ylabel('N', fontsize = 18)
pl.suptitle("Normal mean",fontsize = 18)
```

```python
md['Binomial'] = {} 
 
#and do it in a for loop. not pythonic, but easily readable
n = 0
nn = 0
p = .5  # number of trials, probability of each trial


for n in mysize:
    md['Binomial'][n] = np.random.binomial(2*df, p, n)
    
# save the means for plotting later    

md['Binomial']['means'] = {}
                                           
axbinomial_mu_n = pl.figure(figsize=(10,6)).add_subplot(111)

for nn in md['Binomial']:
    if not type(nn) == str:
        md['Binomial']['means'][nn] = md['Binomial'][nn].mean()
        #and plot it
        axbinomial_mu_n.plot(nn, md['Binomial']['means'][nn], 'o')
        axbinomial_mu_n.set_xlabel('sample size', fontsize = 18)
        axbinomial_mu_n.set_ylabel('sample mean', fontsize = 18)
        axbinomial_mu_n.set_title('Binomial Distribution', fontsize = 18)
        axbinomial_mu_n.plot([min(mysize), max(mysize) + 10], [df, df], 'k')
```

```python
#The bigger the sample the deviation from the mean is smaller
binmean = list(md['Binomial']['means'].values())

pl.figure(figsize=(10, 10))
pl.hist(binmean,bins = 30)
pl.xlabel('sample mean', fontsize = 18)
pl.ylabel('N', fontsize = 18)
pl.suptitle("Binomial mean",fontsize = 18)
```

```python
md['Poisson'] = {} 
 
#and do it in a for loop. not pythonic, but easily readable
n = 0
nn = 0

for n in mysize:
    md['Poisson'][n] = np.random.poisson(df,n)
    
# save the means for plotting later    

md['Poisson']['means'] = {}
                                           
axpoisson_mu_n = pl.figure(figsize = (10,6)).add_subplot(111)

for nn in md['Poisson']:
    if not type(nn) == str:
        md['Poisson']['means'][nn] = md['Poisson'][nn].mean()
        #and plot it
        axpoisson_mu_n.plot(nn, md['Poisson']['means'][nn], 'o')
        axpoisson_mu_n.set_xlabel('sample size', fontsize = 18)
        axpoisson_mu_n.set_ylabel('sample mean', fontsize = 18)
        axpoisson_mu_n.set_title('Poisson Distribution', fontsize = 18)
        axpoisson_mu_n.plot([min(mysize), max(mysize) + 10], [df, df], 'k')
```

```python
#The bigger the sample the deviation from the mean is smaller
posmean = list(md['Poisson']['means'].values())

pl.figure(figsize = (10, 10))
pl.hist(posmean,bins = 30)
pl.xlabel('sample mean', fontsize = 18)
pl.ylabel('N', fontsize = 18)
pl.suptitle("Poisson mean",fontsize = 18)
```

```python
md['Laplace'] = {} 
 
#and do it in a for loop. not pythonic, but easily readable
n = 0
nn = 0
scale = 4
for n in mysize:
    md['Laplace'][n] = np.random.laplace(df,scale,n)
    
# save the means for plotting later    

md['Laplace']['means'] = {}
                                           
axlaplace_mu_n = pl.figure(figsize=(10,6)).add_subplot(111)

for nn in md['Laplace']:
    if not type(nn) == str:
        md['Laplace']['means'][nn] = md['Laplace'][nn].mean()
        #and plot it
        axlaplace_mu_n.plot(nn, md['Laplace']['means'][nn], 'o')
        axlaplace_mu_n.set_xlabel('sample size', fontsize = 18)
        axlaplace_mu_n.set_ylabel('sample mean', fontsize = 18)
        axlaplace_mu_n.set_title('Laplace Distribution', fontsize = 18)
        axlaplace_mu_n.plot([min(mysize), max(mysize) + 10], [df, df], 'k')
```

```python
#The bigger the sample the deviation from the mean is smaller
lapmean = list(md['Laplace']['means'].values())

pl.figure(figsize = (10, 10))
pl.hist(lapmean,bins = 30)
pl.xlabel('sample mean', fontsize = 18)
pl.ylabel('N', fontsize = 18)
pl.suptitle("Laplace mean",fontsize = 18)
```

```python
#themean=np.vstack((lapmean,chimean,normean,binmean,posmean))
l = array(lapmean)
c = array(chimean)
n = array(normean)
b = array(binmean)
p = array(posmean)
lista = [c,n,b,p,l]


#x=sorted(ReprRand.flatten())
pl.figure(figsize = (15, 15))
pl.hist(lista,bins = 17,histtype ='bar', color = ['yellow','blue','magenta','green', 'chartreuse'],\
        label=['ChiSquare', 'Normal', 'Binomial','Poisson','Laplace'])
pl.legend()
pl.suptitle("Distribution of ALL sample means",fontsize = 18)
plt.xlabel('Mean', fontsize = 18)
plt.ylabel('Ocurrance', fontsize = 18)
```

```python
#The central limit theorem can be observed here when plotting the histogram
#all the means for the different distributions and it resemble the Gaussian 
#Distribution.
##https://en.wikipedia.org/wiki/Poisson_distribution
```

```python
#themean=np.vstack((lapmean,chimean,normean,binmean,posmean))
lista=[c,n,b,p,l]

cflat = sorted(c.flatten())
nflat = sorted(n.flatten())
bflat = sorted(b.flatten())
pflat = sorted(p.flatten())
lflat = sorted(l.flatten())

x=[]

x=sorted(cflat+nflat+bflat+pflat+lflat)
             
#plt.scatter(x,c=colors)
adjust=stats.norm.pdf(x,np.mean(x),np.std(x))

#x=sorted(ReprRand.flatten())
pl.figure(figsize=(15, 15))
pl.plot(x,adjust,'-*',c='cyan', label='Normalization', linewidth=5)
pl.hist(lista,bins=17,histtype='bar', color=['yellow','blue','magenta','green', 'chartreuse'], label=['ChiSquare', 'Normal', 'Binomial','Poisson','Laplace'],normed=True)
pl.legend()
pl.suptitle("Distribution of ALL sample means & Normalization",fontsize=18)
plt.xlabel('Mean', fontsize=18)
plt.ylabel('Ocurrance %', fontsize=18)

```
