```python
import os
import sys

# you can use the numpy library for random number generation: 
# http://docs.scipy.org/doc/numpy/reference/routines.random.html

import numpy as np
```

```python
np.random.seed(8)
```

```python
ReprRand=np.random.randn(2,100)
```

```python
ReprRand.shape
```

```python
np.mean(ReprRand)
```

```python
ReprRandAll=np.zeros((50,2,100))
for i in range(50):
    b=np.random.randn(2,100)
    ReprRandAll[i]=(b+i)*i*0.3
   
```

```python
ReprRandAll.shape
```

```python
import pylab as pl
%pylab inline
import scipy.stats as stats
# the last line of code makes the plot show up within the notebook. 
# Always include it so that your homework notebooks are rendered and we can see
# what you wanted your results to be. 

# code here
```

```python
y=np.arange(np.amin(ReprRand),np.amax(ReprRand),.058)
print(len(y))
x=sorted(ReprRand.flatten())
#plt.scatter(x,c=colors)
adjust=stats.norm.pdf(x,np.mean(x),np.std(x))
pl.plot(x,adjust,'-o')
pl.hist(x,normed=True)
pl.suptitle("Normal Distribution randon numbers",fontsize=18)
plt.xlabel('Values for the Random Number', fontsize=18)
plt.ylabel('Ocurrance %', fontsize=18)
plt.show()
```

```python
xmean=np.mean(x)
xstd=np.std(x)
newpdf=stats.norm.pdf(x,xmean,xstd)
pl.plot(x,newpdf,'-o')
plt.show()
print('The mean is:')
print(xmean)
print('The STD is:')
print(xstd)
```
